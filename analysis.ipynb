```python
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data.replace("FALSE", 0, inplace=True)
```

```python
data
```

```python
def resolve_rate(value):
    if value == 'FALSE':
        return 0
    elif type(value) == str:
        return float(value.replace('%', '')) / 100
    else:
        return value
```

```python
for col in data.columns:
    if "Rate" in col:
        data[col] = data[col].apply(resolve_rate)
```
